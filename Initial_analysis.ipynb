```python
import numpy as np
import pandas as pd
import datetime as dt 
import matplotlib.pyplot as plt
from numpy import percentile
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()


```

```python
import io 
df = pd.read_excel(io.BytesIO(uploaded['10949CNov to March.xlsx']))
df['Temperature']=df['Temperature'].round(1)
df

```

```python
# Box plot for week days
sns.set_style("whitegrid")
  
sns.boxplot(x = 'Week_Day', y = 'POWER', data = df)
```

```python
# Load VS Temperature
sns.set_style("whitegrid")
sns.scatterplot(x = 'Temperature', y = 'POWER', data = df)
ax=plt.xlim(-20, 30)
```

```python
# Load vs hour of the day 
sns.set_style("whitegrid")
ax=sns.lineplot(x = 'Hour_day', y = 'POWER', data = df)
ax=plt.xlim(0, 23)
ay=plt.ylim(0, 50)
```

```python
# scatter plot grouped with week day data 
sns.set_style("whitegrid")
sns.relplot( data = df, x ="Temperature", y="POWER", hue ="Week_Day", style="Week_Day",
    kind="scatter" ) 
ax=plt.xlim(-20, 30)
```

```python
#Box plot for Hours of the  day 
sns.set_style("whitegrid")
  
sns.boxplot(x = 'Hour_day', y = 'POWER', data = df)
```
